```python
!pip install scikit-learn

from sklearn.experimental import enable_iterative_imputer

from sklearn.impute import IterativeImputer

numeric_cols_with_missing = df.select_dtypes(include=['float64']).columns[df.select_dtypes(include=['float64']).isnull().sum() > 0]

imputer = IterativeImputer(max_iter=10, random_state=0)

# Fit and transform the data to impute missing values

imputed_data = imputer.fit_transform(df[numeric_cols_with_missing])
# Convert the imputed data back to a DataFrame

imputed_df = pd.DataFrame(imputed_data, columns=numeric_cols_with_missing)
# Combine the imputed numeric columns with the original DataFrame

df[numeric_cols_with_missing] = imputed_df
```

```python
from sklearn.preprocessing import LabelEncoder
labelenc=LabelEncoder()
```

```python
df2['Country'] = labelenc.fit_transform(df2['Country'])
```

```python
df2['Status'] = labelenc.fit_transform(df2['Status'])
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
X = df2
y = df2.pop('Life expectancy ')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=40)
```

```python
!pip install eli5
```

```python
rf = RandomForestRegressor(n_estimators=35)
```

```python
rf.fit(X_train, y_train)
```

```python
from keras.wrappers.scikit_learn import KerasClassifier, KerasRegressor
import eli5
from eli5.sklearn import PermutationImportance

perm = PermutationImportance(rf, random_state=1).fit(X_train,y_train)
eli5.show_weights(perm, feature_names = X_train.columns.tolist())
```

```python
import shap
```

```python
def f_wrapper(X):
    return rf.predict(X).flatten()
```

```python
X_train_summary = shap.kmeans(X_train, 20)

# Compute Shap values
explainer = shap.KernelExplainer(f_wrapper,X_train_summary)

# Make plot with combined shap values
# The training set is too big so let's sample it. We get enough point to draw conclusions
X_train_sample = X_train.sample(400)
shap_values  = explainer.shap_values(X_train_sample)
shap.summary_plot(shap_values, X_train_sample)
```

```python
rf = RandomForestRegressor(n_estimators=5)
```

```python
rf.fit(X_train, y_train)
```

```python
y_pred=rf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test, y_pred)
```
